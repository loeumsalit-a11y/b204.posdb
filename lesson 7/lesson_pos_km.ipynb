## Django គឺជាអ្វី?

Django គឺជា Python web framework កម្រិតខ្ពស់ដែលលើកទឹកចិត្តការអភិវឌ្ឍន៍រហ័ស និងការរចនាស្អាត។ គិតថាវាជា **ឧបករណ៍** — ជំនួសឱ្យការបង្កើតគេហទំព័រពីដំបូង Django ផ្ដល់ឱ្យអ្នកនូវបំណែកដែលបានបង្កើតរួចហើយ ដែលអ្នកអាចភ្ជាប់ចូលគ្នា។

**ឧទាហរណ៍ប្រៀបធៀប**: ការបង្កើតប្រព័ន្ធ POS ដោយគ្មាន Django គឺដូចជាការភ្ជាប់ម៉ាស៊ីនគិតប្រាក់ពីសូន្យដោយប្រើសមាសធាតុអេឡិចត្រូនិក។ Django ផ្ដល់ឱ្យអ្នកនូវអេក្រង់ ក្តារចុច និងម៉ាស៊ីនបោះពុម្ពវិក្កយបត្ររួចជាស្រេច — អ្នកគ្រាន់តែកំណត់រចនាសម្ព័ន្ធវាសម្រាប់ហាងរបស់អ្នក។

លក្ខណៈពិសេសសំខាន់ៗរបស់ Django:
- **ការអភិវឌ្ឍន៍រហ័ស**: ឧបករណ៍ built-in សម្រាប់ authentication ការគ្រប់គ្រងទិន្នន័យ និងផ្ទាំង admin
- **ការពង្រីក**: ប្រើដោយ **Instagram**, **Pinterest**, និង **Disqus** ដើម្បីបម្រើអ្នកប្រើប្រាស់រាប់លាននាក់
- **សុវត្ថិភាព**: ការពារដោយស្វ័យប្រវត្តិប្រឆាំងនឹង SQL injection, XSS, និង CSRF
- **ភាពចម្រុះ**: សមរម្យសម្រាប់អ្វីៗចាប់ពីហាងមូលដ្ឋានតូចៗ រហូតដល់ហាងធំៗ

## ហេតុអ្វីត្រូវប្រើ Django?

| លក្ខណៈពិសេស | អត្ថប្រយោជន៍ |
|---|---|
| **Batteries Included** | ផ្ទាំង Admin, authentication, ORM — ផ្ដល់ built-in ទាំងអស់ |
| **MVT Architecture** | ការបំបែកច្បាស់លាស់ នៃទិន្នន័យ (Model), តក្កវិទ្យា (View) និងការរចនា (Template) |
| **គោលការណ៍ DRY** | "កុំធ្វើអ្វីម្ដងហើយម្ដងទៀត" — សរសេរកូដម្ដង ប្រើឡើងវិញគ្រប់ទីកន្លែង |
| **ឯកសារប្រើប្រាស់ល្អ** | Framework Python ដែលមានឯកសារប្រើប្រាស់ល្អបំផុត |
| **សហគមន៍សកម្ម** | ប្រព័ន្ធអេកូដ៏ធំ plugin និង package នៅ [PyPI](https://pypi.org/) |

### ការពន្យល់ MVT Pattern របស់ Django

```
សំណើអ្នកប្រើ → URL Router → View (តក្កវិទ្យា)
                         ↓          ↑
                        Model      Template
                        (ទិន្នន័យ)    (HTML)
```

សម្រាប់ប្រព័ន្ធ POS របស់យើង:
- **Model** → តារាង `Product`, `Order`, `OrderItem` នៅក្នុងមូលដ្ឋានទិន្នន័យ
- **View** → functions ដែលទាញផលិតផល/ការបញ្ជាទិញ ហើយបញ្ជូនទៅ templates
- **Template** → ទំព័រ HTML ដែលបុគ្គលិកឃើញនៅលើអេក្រង់

## ស្វែងយល់ពីរចនាសម្ព័ន្ធគម្រោង Django

នេះជាអ្វីដែល Django បង្កើតដោយស្វ័យប្រវត្តិ:

posdb/                    ← ថតឫស (គម្រោងរបស់អ្នក)
    manage.py             ← ឧបករណ៍បន្ទាត់ command (run server, migrations ជាដើម)
    posdb/                ← "configuration package" របស់គម្រោង
        __init__.py
        settings.py       ← ការកំណត់គម្រោងទាំងអស់ (database, apps, time zone…)
        urls.py           ← URL dispatcher កំពូល
        asgi.py           ← ចំណុចចូលម៉ាស៊ីនបម្រើ async
        wsgi.py           ← ចំណុចចូលម៉ាស៊ីនបម្រើប្រពៃណី

| ឯកសារ | គោលបំណង |
|---|---|
| `manage.py` | ឧបករណ៍ចម្បងរបស់អ្នក — run server, បង្កើត migrations, បើក shell |
| `settings.py` | កំណត់រចនាសម្ព័ន្ធ databases, installed apps, time zone, static files |
| `urls.py` | "ទូរទស្សន៍ reception" — បញ្ជូន URLs ចូលទៅ view ត្រឹមត្រូវ |
| `wsgi.py` / `asgi.py` | ប្រើនៅពេល deploy ទៅ production server |

**តេស្តរហ័ស**: ចាប់ផ្ដើម development server ហើយចូល `http://127.0.0.1:8000/`

```bash
python manage.py runserver
```

## គំរូ (Models) និង Django ORM

**Model** គឺជា Python class ដែលតំណាងឱ្យ database table។ attribute នីមួយៗគឺជាជួរឈរ (column)។

**ORM (Object-Relational Mapper)** របស់ Django អនុញ្ញាតឱ្យអ្នក:
- បង្កើត អាន ធ្វើបច្ចុប្បន្នភាព និងលុប records ដោយប្រើ Python — មិនត្រូវការ SQL ទេ
- បង្កើត database migrations ដោយស្វ័យប្រវត្តិ នៅពេលអ្នកផ្លាស់ប្ដូរ model

### គំរូទិន្នន័យ POS របស់យើង

ប្រព័ន្ធ POS ធម្មតាមាន tables ចម្បងបី។ ":< " មានន័យថា "មានច្រើន":

```
Product ──< OrderItem >── Order
```

- **Product**: ទំនិញដែលហាងលក់ (ឈ្មោះ, តម្លៃ, ស្តុក)
- **Order**: ប្រតិបត្តិការលក់មួយ (អ្នកគិតលុយ, ពេលវេលា, សរុប)
- **OrderItem**: បន្ទាត់មួយនៅលើវិក្កយបត្រ — ផលិតផលអ្វី, ចំនួនប៉ុន្មាន, ក្នុងតម្លៃប៉ុន្មាន

### ការកំណត់ Models

បើក `sales/models.py` ហើយកំណត់ models ទាំងបី:

````python
# sales/models.py

from django.db import models


class Product(models.Model):
    CATEGORY_CHOICES = [
        ('food',        'អាហារ និងភេសជ្ជៈ'),
        ('electronics', 'អេឡិចត្រូនិក'),
        ('clothing',    'សម្លៀកបំពាក់'),
        ('household',   'គ្រឿងសង្ហារឹម'),
        ('other',       'ផ្សេងៗ'),
    ]

    name       = models.CharField(max_length=200)
    category   = models.CharField(max_length=20, choices=CATEGORY_CHOICES)
    price      = models.DecimalField(max_digits=8, decimal_places=2)   # ឧ. 12.99
    stock      = models.PositiveIntegerField(default=0)                # ចំនួនក្នុងស្តុក
    barcode    = models.CharField(max_length=50, unique=True, blank=True)
    is_active  = models.BooleanField(default=True)                     # លាក់ទំនិញឈប់លក់

    def __str__(self):
        return f"{self.name}  —  ${self.price}  (ស្តុក: {self.stock})"

    class Meta:
        ordering = ['name']


class Order(models.Model):
    STATUS_CHOICES = [
        ('open',       'បានបើក'),
        ('paid',       'បានបង់'),
        ('refunded',   'បានសង'),
        ('cancelled',  'បានលុបចោល'),
    ]

    cashier    = models.CharField(max_length=100)           # ឈ្មោះ ឬ ID បុគ្គលិក
    status     = models.CharField(max_length=20, choices=STATUS_CHOICES, default='open')
    created_at = models.DateTimeField(auto_now_add=True)    # កំណត់ពេលបង្កើតការបញ្ជាទិញ
    notes      = models.TextField(blank=True)

    @property
    def total(self):
        """បូករួម subtotal នៃ line item នីមួយៗក្នុងការបញ្ជាទិញនេះ"""
        return sum(item.subtotal for item in self.items.all())

    def __str__(self):
        return f"ការបញ្ជាទិញ #{self.pk}  [{self.status.upper()}]  —  ${self.total:.2f}"

    class Meta:
        ordering = ['-created_at']


class OrderItem(models.Model):
    order      = models.ForeignKey(Order, on_delete=models.CASCADE, related_name='items')
    product    = models.ForeignKey(Product, on_delete=models.PROTECT)   # PROTECT ការពារការលុបផលិតផលដែលមានការលក់
    quantity   = models.PositiveIntegerField(default=1)
    unit_price = models.DecimalField(max_digits=8, decimal_places=2)    # តម្លៃនៅពេលលក់

    @property
    def subtotal(self):
        return self.unit_price * self.quantity

    def __str__(self):
        return f"{self.quantity} × {self.product.name}  @  ${self.unit_price}"
````

### គំនិត ORM សំខាន់ៗដែលបានប្រើខាងលើ

| គំនិត | ឧទាហរណ៍ | ន័យ |
|---|---|---|
| `ForeignKey` | `order = ForeignKey(Order, ...)` | OrderItems ជាច្រើន ជាប់ទាក់ទងនឹង Order មួយ |
| `on_delete=CASCADE` | Order ត្រូវបានលុប → OrderItems របស់វាត្រូវបានលុបផង |
| `on_delete=PROTECT` | មិនអាចលុប Product ដែលបង្ហាញក្នុង order បានទេ |
| `related_name='items'` | `order.items.all()` — ចូលប្រើ items ពីខាង Order |
| `@property` | `order.total` — តម្លៃគណនា មិន store ក្នុង DB |

### ការ Run Migrations

````python
# ជំហានទី 1: បង្កើតឯកសារ migration
python manage.py makemigrations sales

# ជំហានទី 2: អនុវត្តទៅ database
python manage.py migrate

# លទ្ធផលរំពឹងទុក:
#   Applying sales.0001_initial... OK
````

## ទិដ្ឋភាព (Views) និង ទំព័រគំរូ (Templates)

### បង្កើត Views

**View** គឺជា Python function ដែល:
1. ទទួល HTTP request
2. Query database (តាមរយៈ Model)
3. បញ្ជូនទិន្នន័យទៅ template
4. Return HTTP response (HTML ដែល render)

បើក `sales/views.py` ហើយសរសេរ views បី:

````python
# sales/views.py

from django.shortcuts import render, get_object_or_404
from .models import Product, Order


def product_list(request):
    """បង្ហាញផលិតផល active ទាំងអស់ តម្រៀប A–Z"""
    products = Product.objects.filter(is_active=True)
    return render(request, 'sales/product_list.html', {'products': products})


def product_detail(request, pk):
    """បង្ហាញព័ត៌មានលម្អិតសម្រាប់ផលិតផលតែមួយ។ Return 404 ប្រសិនបើរកមិនឃើញ"""
    product = get_object_or_404(Product, pk=pk)
    return render(request, 'sales/product_detail.html', {'product': product})


def order_list(request):
    """បង្ហាញការបញ្ជាទិញទាំងអស់ ថ្មីបំផុតមុន"""
    orders = Order.objects.all()
    return render(request, 'sales/order_list.html', {'orders': orders})
````

## ការបញ្ជូន URL ក្នុង Django

URL routing ភ្ជាប់ URL pattern (ដូចជា `/sales/products/3/`) ទៅ view function ដែលដំណើរការវា។

### ជំហានទី 1 — URLs របស់ App (`sales/urls.py`)

បង្កើតឯកសារថ្មី `sales/urls.py`:

````python
# sales/urls.py

from django.urls import path
from . import views

urlpatterns = [
    # /sales/products/       → បញ្ជីផលិតផល
    path('products/', views.product_list, name='product_list'),

    # /sales/products/1/     → ទំព័រលម្អិតសម្រាប់ product id=1
    path('products/<int:pk>/', views.product_detail, name='product_detail'),

    # /sales/orders/         → ការបញ្ជាទិញទាំងអស់
    path('orders/', views.order_list, name='order_list'),
]
````

### ជំហានទី 2 — URLs របស់គម្រោង (`posdb/urls.py`)

ភ្ជាប់ URLs របស់ app ចូលក្នុងឯកសារ URL គម្រោងសំខាន់ ដោយប្រើ `include()`:

````python
# posdb/urls.py

from django.contrib import admin
from django.urls import path, include

urlpatterns = [
    path('admin/', admin.site.urls),          # /admin/  → ផ្ទាំង Django admin
    path('sales/', include('sales.urls')),    # /sales/  → បញ្ជូនទៅ sales/urls.py
]

# ឧទាហរណ៍លំហូរ URL:
# Request: GET /sales/products/3/
# ផ្គូផ្គង: path('sales/', ...) → កាត់ 'sales/' ហើយបញ្ជូន 'products/3/' ទៅ sales/urls.py
# ផ្គូផ្គង: path('products/<int:pk>/', ...) → ហៅ product_detail(request, pk=3)
````

## ការប្រើប្រាស់ Django Admin

ផ្ទាំង admin របស់ Django គឺជា **dashboard ដែលបង្កើតដោយស្វ័យប្រវត្តិ ហើយឥតគិតថ្លៃ** សម្រាប់គ្រប់គ្រងទិន្នន័យ — ល្អឥតខ្ចោះសម្រាប់អ្នកគ្រប់គ្រងហាងដើម្បីបន្ថែមផលិតផល មើលការបញ្ជាទិញ និងធ្វើបច្ចុប្បន្នភាពស្តុក។

### ជំហានទី 1 — ចុះឈ្មោះ Models

បើក `sales/admin.py` ហើយចុះឈ្មោះ models ទាំងបី:

````python
# sales/admin.py

from django.contrib import admin
from .models import Product, Order, OrderItem


@admin.register(Product)
class ProductAdmin(admin.ModelAdmin):
    list_display  = ['name', 'category', 'price', 'stock', 'is_active']
    list_filter   = ['category', 'is_active']
    search_fields = ['name', 'barcode']
    ordering      = ['name']


class OrderItemInline(admin.TabularInline):
    """បង្ហាញ order items ផ្ទាល់នៅក្នុងទំព័រកែ Order"""
    model  = OrderItem
    extra  = 1    # ចំនួនជួរទទេដែលបង្ហាញសម្រាប់បន្ថែម items ថ្មី
    fields = ['product', 'quantity', 'unit_price']


@admin.register(Order)
class OrderAdmin(admin.ModelAdmin):
    list_display  = ['pk', 'cashier', 'status', 'created_at']
    list_filter   = ['status']
    search_fields = ['cashier', 'notes']
    ordering      = ['-created_at']
    inlines       = [OrderItemInline]    # ← បង្ហាញ items នៅក្នុងទម្រង់ order
````

````python
# ── Run these lines inside: python manage.py shell ──

from sales.models import Product, Order, OrderItem

# ── 1. បង្កើតផលិតផល ──────────────────────────────────────────
coffee = Product.objects.create(
    name='កាហ្វេ Espresso',
    category='food',
    price=1.50,
    stock=100,
    barcode='FOOD-001',
)
sandwich = Product.objects.create(
    name='នំប៉័ង Club',
    category='food',
    price=3.99,
    stock=20,
    barcode='FOOD-002',
)
headphones = Product.objects.create(
    name='កាស Wireless',
    category='electronics',
    price=49.99,
    stock=15,
    barcode='ELEC-001',
)
charger = Product.objects.create(
    name='ខ្សែសាក USB-C',
    category='electronics',
    price=9.99,
    stock=4,      # ស្តុកទាប!
    barcode='ELEC-002',
)

print("ផលិតផលបានបង្កើត!")

# ── 2. បង្កើតការបញ្ជាទិញជាមួយ Items ──────────────────────────
order1 = Order.objects.create(cashier='សុភា', status='paid')

OrderItem.objects.create(
    order=order1,
    product=coffee,
    quantity=2,
    unit_price=coffee.price,
)
OrderItem.objects.create(
    order=order1,
    product=sandwich,
    quantity=1,
    unit_price=sandwich.price,
)

print(f"សរុបការបញ្ជាទិញ: ${order1.total:.2f}")   # គួរតែ print $6.99

# ── 3. ឧទាហរណ៍ Query ──────────────────────────────────────────

# ផលិតផលទាំងអស់ តម្រៀប A-Z
all_products = Product.objects.all()
print(f"ផលិតផលសរុប: {all_products.count()}")

# តែទំនិញ food
food_items = Product.objects.filter(category='food')
for p in food_items:
    print(p)

# ការព្រមានស្តុកទាប: ផលិតផលដែលមាន 5 ឬតិចជាងនេះ
low_stock = Product.objects.filter(stock__lte=5)
print("\n⚠️  ទំនិញស្តុកទាប:")
for p in low_stock:
    print(f"  {p.name}: {p.stock} នៅសល់")

# ផលិតផលតម្លៃលើសពី $10
expensive = Product.objects.filter(price__gt=10).order_by('-price')
for p in expensive:
    print(f"{p.name}: ${p.price}")

# ── 4. ធ្វើបច្ចុប្បន្នភាពស្តុក ────────────────────────────────
charger = Product.objects.get(barcode='ELEC-002')
charger.stock += 50    # ទំនិញផ្គត់ផ្គង់មកដល់ហើយ
charger.save()
print(f"ស្តុកបានធ្វើបច្ចុប្បន្នភាព {charger.name}: {charger.stock}")

# ── 5. Items ទាំងអស់នៅក្នុងការបញ្ជាទិញ #1 ────────────────────
for item in order1.items.all():
    print(f"  {item}  →  subtotal: ${item.subtotal:.2f}")
````

**តារាងសង្ខេប ORM Query**

| គោលដៅ | ORM Query |
|---|---|
| Records ទាំងអស់ | `Product.objects.all()` |
| Filter តាម field | `Product.objects.filter(category='food')` |
| Exclude records | `Product.objects.exclude(is_active=False)` |
| តិចជាងឬស្មើ | `Product.objects.filter(stock__lte=5)` |
| ធំជាង | `Product.objects.filter(price__gt=10)` |
| តម្រៀបលទ្ធផល | `.order_by('-price')` (prefix `-` = ចុះ) |
| Get records មួយ | `Product.objects.get(barcode='FOOD-001')` |
| Get ឬ 404 | `get_object_or_404(Product, pk=1)` |
| រាប់ | `Order.objects.count()` |
| Follow FK ចេញ | `order.items.all()` — OrderItems ទាំងអស់សម្រាប់ order |
| Follow FK ចូល | `item.product.name` — ផលិតផលលើ OrderItem |

### កិច្ចការទី 6 — ការប្រឡងបន្ថែម ⭐

បន្ថែម model `Discount` ដែលអាចអនុវត្តទៅការបញ្ជាទិញ:

```python
class Discount(models.Model):
    order       = models.OneToOneField(Order, on_delete=models.CASCADE, related_name='discount')
    description = models.CharField(max_length=200)   # ឧ. "ការបញ្ចុះតម្លៃបុគ្គលិក"
    amount      = models.DecimalField(max_digits=6, decimal_places=2)

    def __str__(self):
        return f"{self.description} (—${self.amount}) នៅការបញ្ជាទិញ #{self.order.pk}"
```

បន្ទាប់មក:
1. Run migrations សម្រាប់ model ថ្មី។
2. ចុះឈ្មោះ `Discount` ក្នុង `admin.py`។
3. ធ្វើបច្ចុប្បន្នភាព `Order.total` ដើម្បីដក discount amount ប្រសិនបើមួយមាន:
   ```python
   @property
   def total(self):
       subtotal = sum(item.subtotal for item in self.items.all())
       try:
           return subtotal - self.discount.amount
       except Discount.DoesNotExist:
           return subtotal
   ```
4. តេស្តវានៅ shell ដោយបង្កើត `Discount` សម្រាប់ order ដែលមានស្រាប់ ហើយផ្ទៀងផ្ទាត់ `order.total` ផ្លាស់ប្ដូរ។
